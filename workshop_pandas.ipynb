## Objetivo
Resolver un caso real de negocio trabajando con datos sucios, inconsistentes y provenientes de múltiples fuentes.

Trabajaremos con:
- clientes.csv
- ventas.csv
- productos.csv

```python

import pandas as pd
```

## 1. Cargar datos

```python

clientes = pd.read_csv('clientes.csv')
ventas = pd.read_csv('ventas.csv')
productos = pd.read_csv('productos.csv')

clientes.head()
```

## 2. Exploración inicial

```python

clientes.info()
ventas.info()
productos.info()
```

## 3. Diagnóstico de calidad de datos

```python

clientes.isnull().sum()
```

## 4. Limpieza de clientes

```python

# Normalizar nombres
clientes['nombre'] = clientes['nombre'].str.title()

# Normalizar ciudades
clientes['ciudad'] = clientes['ciudad'].str.lower().str.strip()

# Convertir fechas
clientes['fecha_registro'] = pd.to_datetime(clientes['fecha_registro'], errors='coerce')

# Manejar nulos en edad
clientes['edad'] = clientes['edad'].fillna(clientes['edad'].median())

clientes.head()
```

## 5. Limpieza de ventas

```python

# Limpiar precios
ventas['precio'] = ventas['precio'].astype(str).str.replace('$','', regex=False)
ventas['precio'] = pd.to_numeric(ventas['precio'], errors='coerce')

# Filtrar cantidades inválidas
ventas = ventas[ventas['cantidad'] > 0]

# Convertir fechas
ventas['fecha_venta'] = pd.to_datetime(ventas['fecha_venta'], errors='coerce')

ventas.head()
```

## 6. Merge de datos

```python

df = ventas.merge(clientes, on='cliente_id', how='left')
df.head()
```

```python

df.isnull().sum()
```

## 7. Validación con productos

```python

df = df.merge(productos, on='producto', how='left')
df.head()
```

## 8. Métricas de negocio

```python

df['ingreso'] = df['precio'] * df['cantidad']

df.groupby('producto')['ingreso'].sum().sort_values(ascending=False)
```

```python
import pandas as pd

# Ejemplo: columna con fechas sucias
df = pd.DataFrame({
    "fecha": ["2023-01-15", "15/02/2023", "03-20-2023", "2023/04/10", "fecha_invalida"]
})

# Convertir a datetime (maneja errores)
df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")

# Ver resultados
print(df)
```

```python
# Verificar nulos 
# Detectar fechas nulas
df[df["fecha"].isnull()]
```

```python
df["año"] = df["fecha"].dt.year
df["mes"] = df["fecha"].dt.month
df["dia"] = df["fecha"].dt.day
df["dia_semana"] = df["fecha"].dt.day_name()
```

```python
df["trimestre"] = df["fecha"].dt.quarter
```

```python
df["fecha_sin_hora"] = df["fecha"].dt.date
```

```python
df["fecha_formateada"] = df["fecha"].dt.strftime("%Y-%m")
```
